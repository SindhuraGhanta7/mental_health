```python
import pandas as pd
import numpy as np
import re

# --- Load ---
df = pd.read_csv("Mental_Health_Dataset[1].csv")
display(df)
```

```python
for c in df.select_dtypes(include=["object"]).columns:
    df[c] = df[c].astype(str).str.strip()

# --- Timestamp ---
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")
display(df)
```

```python
def norm_yes_no_maybe(s: pd.Series) -> pd.Series:
    mapping = {
        "y": "Yes", "yes": "Yes", "ye": "Yes", "yep": "Yes", "true": "Yes", "1": "Yes",
        "n": "No", "no": "No", "nah": "No", "false": "No", "0": "No",
        "maybe": "Maybe", "na": "Maybe", "n/a": "Maybe", "not sure": "Maybe", "unsure": "Maybe", "unknown": "Maybe"
    }
    s_norm = s.str.lower().map(lambda x: re.sub(r"[^a-z0-9+\-\/ ]+", "", x))  # keep alnum and simple tokens
    return s_norm.replace(mapping).fillna(s)

def set_ordered_cat(s: pd.Series, order: list) -> pd.Series:
    return pd.Categorical(s, categories=order, ordered=True)
```

```python
def normalize_gender(val: str) -> str:
    v = val.strip().lower()
    if v in {"m", "male", "man", "cis male", "cis-man", "cis man"}:
        return "Male"
    if v in {"f", "female", "woman", "cis female", "cis-female", "cis woman"}:
        return "Female"
    if v in {"non-binary", "nonbinary", "nb", "genderqueer", "other", "trans", "transgender", "agender", "prefer not to say", "na", "n/a"}:
        return "Other / Prefer not to say"
    # catch shorthands
    if v.startswith("m "): return "Male"
    if v.startswith("f "): return "Female"
    return "Other / Prefer not to say"

df["Gender"] = df["Gender"].apply(normalize_gender)
display(df)
```

```python
country_map = {
    "us": "United States",
    "usa": "United States",
    "u.s.a": "United States",
    "united states of america": "United States",
    "u.s.": "United States",
    "uk": "United Kingdom",
    "u.k.": "United Kingdom",
    "england": "United Kingdom",
    "scotland": "United Kingdom",
    "wales": "United Kingdom",
    "uae": "United Arab Emirates",
    "korea, republic of": "South Korea",
    "korea, south": "South Korea",
    "korea": "South Korea",
    "russia": "Russian Federation",
    "viet nam": "Vietnam",
}
def normalize_country(val: str) -> str:
    v = val.strip().lower()
    return country_map.get(v, val.title().strip())

df["Country"] = df["Country"].apply(normalize_country)
```

```python
df["self_employed"] = df["self_employed"].replace({"": np.nan})
df["self_employed"] = df["self_employed"].fillna("Unknown")
```

```python
yes_no_maybe_cols = [
    "family_history", "treatment", "Growing_Stress", "Changes_Habits",
    "Mental_Health_History", "Coping_Struggles", "Work_Interest",
    "Social_Weakness", "mental_health_interview", "care_options"
]

for c in yes_no_maybe_cols:
    if c in df.columns:
        df[c] = norm_yes_no_maybe(df[c])
        # Limit to canonical set if possible
        df[c] = df[c].where(df[c].isin(["Yes", "No", "Maybe"]), "Maybe")
```

```python
def norm_days_indoors(x: str) -> str:
    v = x.strip().lower()
    v = v.replace("day(s)", "days")
    if re.search(r"^(1|01)\s*-\s*(14|14)\s*days$", v) or "1-14" in v:
        return "1-14 days"
    if "15-30" in v or re.search(r"^15\s*-\s*30", v):
        return "15-30 days"
    if "more than 30" in v or "30+" in v or "over 30" in v:
        return "More than 30 days"
    if "everyday" in v or "every day" in v or v == "daily":
        return "Everyday"
    if "rare" in v or v in {"rarely"}:
        return "Rarely"
    return x.strip()

if "Days_Indoors" in df.columns:
    df["Days_Indoors"] = df["Days_Indoors"].apply(norm_days_indoors)
    df["Days_Indoors"] = set_ordered_cat(
        df["Days_Indoors"],
        ["Rarely", "1-14 days", "15-30 days", "More than 30 days", "Everyday"]
    )
```

```python
def norm_mood(x: str) -> str:
    v = x.strip().lower()
    if v.startswith("low"): return "Low"
    if v.startswith("med"): return "Medium"
    if v.startswith("high"): return "High"
    if v in {"0", "1"}: return "Low"
    if v in {"2"}: return "Medium"
    if v in {"3"}: return "High"
    return "Medium"

if "Mood_Swings" in df.columns:
    df["Mood_Swings"] = df["Mood_Swings"].apply(norm_mood)
    df["Mood_Swings"] = set_ordered_cat(df["Mood_Swings"], ["Low", "Medium", "High"])
```

```python
if "Occupation" in df.columns:
    df["Occupation"] = df["Occupation"].replace({"": np.nan})
    df["Occupation"] = df["Occupation"].fillna("Unknown").str.title().str.replace(r"\s+", " ", regex=True)
```

```python
df = df.drop_duplicates().reset_index(drop=True)

# --- Enforce dtypes for clean categories ---
for c in yes_no_maybe_cols:
    if c in df.columns:
        df[c] = pd.Categorical(df[c], categories=["No", "Maybe", "Yes"], ordered=True)
```

```python
def bin_map(s: pd.Series) -> pd.Series:
    return s.map({"No": 0, "Yes": 1}).astype("Int64")

binary_targets = ["family_history", "treatment", "Coping_Struggles", "mental_health_interview", "care_options", "Work_Interest", "Social_Weakness", "Mental_Health_History"]
for c in binary_targets:
    if c in df.columns:
        df[f"{c}_bin"] = bin_map(df[c])

if "Mood_Swings" in df.columns:
    df["Mood_Swings_ord"] = df["Mood_Swings"].map({"Low": 0, "Medium": 1, "High": 2}).astype("Int64")

if "Days_Indoors" in df.columns:
    df["Days_Indoors_ord"] = df["Days_Indoors"].map({
        "Rarely": 0, "1-14 days": 1, "15-30 days": 2, "More than 30 days": 3, "Everyday": 4
    }).astype("Int64")
```

```python
placeholders = {"nan", "none", "null", "?", "-", "--"}
for c in df.select_dtypes(include=["object"]).columns:
    df[c] = df[c].apply(lambda x: np.nan if str(x).strip().lower() in placeholders else x)

# Keep self_employed as non-null categorical
df["self_employed"] = df["self_employed"].fillna("Unknown")
```

```python
df.to_csv("Mental_Health_Dataset_cleaned.csv", index=False)
display(df.head())
print(df.shape)
```

```python
# Visualization 1 — Monthly Treatment Rate Over Time (color-coded line)
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

dft = df.copy()
dft["Timestamp"] = pd.to_datetime(dft["Timestamp"], errors="coerce")
dft = dft.dropna(subset=["Timestamp"])
dft["month"] = dft["Timestamp"].dt.to_period("M").dt.to_timestamp()

def _ynm(s: pd.Series) -> pd.Series:
    m = {
        "y":"Yes","yes":"Yes","true":"Yes","1":"Yes",
        "n":"No","no":"No","false":"No","0":"No",
        "maybe":"Maybe","na":"Maybe","n/a":"Maybe","not sure":"Maybe","unsure":"Maybe","unknown":"Maybe"
    }
    return s.astype(str).str.strip().str.lower().replace(m)

dft["treatment"] = _ynm(dft["treatment"])
mask = dft["treatment"].isin(["Yes","No"])
by_month = (
    dft[mask]
    .groupby("month")["treatment"]
    .apply(lambda s: (s == "Yes").mean())
    .reset_index(name="treatment_rate")
    .sort_values("month")
)

plt.figure(figsize=(11, 4))
plt.plot(
    by_month["month"], by_month["treatment_rate"],
    color="#2E86AB", linewidth=2.5, marker="o", markersize=3, alpha=0.95
)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1.0))
plt.title("Monthly Treatment Rate Over Time")
plt.xlabel("Month")
plt.ylabel("Treatment Rate")
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
# Visualization 2 — Top 15 Countries by Respondent Count (color-scaled bars)
import matplotlib.pyplot as plt
from matplotlib import cm, colors

country_counts = (
    df["Country"].astype(str).str.strip()
    .value_counts()
    .head(15)
    .sort_values()
)

vals = country_counts.values.astype(float)
norm = colors.Normalize(vmin=vals.min(), vmax=vals.max())
cmap = cm.get_cmap("Blues")  # colorblind-friendly sequential palette
bar_colors = cmap(norm(vals))

plt.figure(figsize=(9, 6))
plt.barh(country_counts.index.astype(str), vals, color=bar_colors, edgecolor="#1b2838", linewidth=0.6)
for i, v in enumerate(vals):
    plt.text(v, i, f" {int(v)}", va="center", fontsize=10)
plt.title("Top 15 Countries by Respondent Count")
plt.xlabel("Number of Respondents")
plt.ylabel("Country")
plt.grid(axis="x", alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# Visualization 3 — Mood Swings by Days Indoors (100% Stacked, color-coded categories)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfx = df.copy()

def _norm_mood(x: str) -> str:
    v = str(x).strip().lower()
    if v.startswith("low"): return "Low"
    if v.startswith("med"): return "Medium"
    if v.startswith("high"): return "High"
    return "Medium"

def _norm_days(x: str) -> str:
    v = str(x).strip().lower().replace("day(s)", "days")
    if "1-14" in v: return "1-14 days"
    if "15-30" in v: return "15-30 days"
    if "more than 30" in v or "30+" in v or "over 30" in v: return "More than 30 days"
    if "everyday" in v or "every day" in v or v == "daily": return "Everyday"
    if "rare" in v or v == "rarely": return "Rarely"
    return str(x).strip()

dfx["Mood_Swings"] = dfx["Mood_Swings"].apply(_norm_mood)
dfx["Days_Indoors"] = dfx["Days_Indoors"].apply(_norm_days)

cat_days = ["Rarely","1-14 days","15-30 days","More than 30 days","Everyday"]
cat_mood = ["Low","Medium","High"]
dfx = dfx[dfx["Days_Indoors"].isin(cat_days) & dfx["Mood_Swings"].isin(cat_mood)].copy()
dfx["Days_Indoors"] = pd.Categorical(dfx["Days_Indoors"], categories=cat_days, ordered=True)
dfx["Mood_Swings"]   = pd.Categorical(dfx["Mood_Swings"],   categories=cat_mood, ordered=True)

ct = pd.crosstab(dfx["Days_Indoors"], dfx["Mood_Swings"]).reindex(index=cat_days, columns=cat_mood).fillna(0)
ct_pct = ct.div(ct.sum(axis=1), axis=0)

colors_mood = {
    "Low":   "#4C78A8",  # blue
    "Medium":"#F2CF5B",  # gold
    "High":  "#E45756"   # red
}

plt.figure(figsize=(10, 5))
x = np.arange(len(ct_pct.index))
bottom = np.zeros(len(ct_pct.index))
for mood in cat_mood:
    vals = ct_pct[mood].values
    plt.bar(x, vals, bottom=bottom, label=mood, color=colors_mood[mood], edgecolor="white", linewidth=0.7)
    bottom += vals

plt.title("Mood Swings Distribution by Days Indoors (Row-Normalized)")
plt.xlabel("Days Indoors")
plt.ylabel("Proportion")
plt.xticks(x, ct_pct.index)
plt.ylim(0, 1)
plt.grid(axis="y", alpha=0.25)
plt.legend(title="Mood Swings", ncol=3, frameon=False)
plt.tight_layout()
plt.show()
```

```python
# Visualization 4 (annotated) — Correlation Heatmap with numbers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfh = df.copy()

def _bin(s: pd.Series) -> pd.Series:
    return s.map({"No":0, "Yes":1}).astype("float")

def _yn(s: pd.Series) -> pd.Series:
    return s.astype(str).str.strip().str.title().replace({"Yes":"Yes", "No":"No"})

bin_cols_src = [
    "family_history","treatment","Coping_Struggles",
    "mental_health_interview","care_options","Work_Interest",
    "Social_Weakness","Mental_Health_History"
]
num_cols = []

for c in bin_cols_src:
    if c in dfh.columns:
        dfh[c] = _yn(dfh[c])
        dfh[c + "_bin"] = _bin(dfh[c])
        num_cols.append(c + "_bin")

if "Mood_Swings" in dfh.columns:
    dfh["Mood_Swings"] = dfh["Mood_Swings"].astype(str).str.strip().str.title()
    dfh["Mood_Swings_ord"] = dfh["Mood_Swings"].map({"Low":0,"Medium":1,"High":2}).astype("float")
    num_cols.append("Mood_Swings_ord")

if "Days_Indoors" in dfh.columns:
    di = dfh["Days_Indoors"].astype(str).str.strip().str.title()
    di = di.replace({
        "1-14":"1-14 days","15-30":"15-30 days",
        "30+":"More than 30 days","Over 30":"More than 30 days","Daily":"Everyday"
    })
    dmap = {"Rarely":0,"1-14 Days":1,"1-14 days":1,"15-30 Days":2,"15-30 days":2,"More Than 30 Days":3,"More Than 30 days":3,"Everyday":4}
    dfh["Days_Indoors_ord"] = di.map(dmap).astype("float")
    num_cols.append("Days_Indoors_ord")

corr_df = dfh[num_cols].select_dtypes(include=[np.number]).dropna(axis=1, how="all")
corr = corr_df.corr()

fig, ax = plt.subplots(figsize=(9, 8))
im = ax.imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)

# Colorbar
cbar = plt.colorbar(im, fraction=0.046, pad=0.04)
cbar.set_label("Pearson r", rotation=270, labelpad=15)

# Axis ticks/labels
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.index)))
ax.set_xticklabels(corr.columns, rotation=90)
ax.set_yticklabels(corr.index)
ax.set_title("Correlation Heatmap of Mental-Health Indicators")

# Optional: fine gridlines for readability
ax.set_xticks(np.arange(-.5, len(corr.columns), 1), minor=True)
ax.set_yticks(np.arange(-.5, len(corr.index), 1), minor=True)
ax.grid(which="minor", color="white", linestyle="-", linewidth=0.5, alpha=0.5)
ax.tick_params(which="minor", bottom=False, left=False)

# --- Annotate each cell with value ---
# White text on strong correlations, dark text on weaker ones
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        val = corr.iloc[i, j]
        if np.isnan(val):
            continue
        txt_color = "white" if abs(val) >= 0.5 else "#222222"
        ax.text(j, i, f"{val:.2f}", ha="center", va="center", color=txt_color, fontsize=9)

plt.tight_layout()
plt.show()
```
